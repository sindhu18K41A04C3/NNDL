```
import numpy as np
import pandas as pd
d = pd.read_csv('/content/pressure and weight.csv')
#print(d)
x=np.array(d.iloc[:,0:1])
y=np.array(d.iloc[:,1:2])
x_min = x.min()
x_max = x.max()
y_min = y.min()
y_max = y.max()
#normalizing the given dataset
for i in range(len(x)):
  x[i] = (x[i]-x.min())/(x.max()-x.min())
  y[i] = (y[i]-y.min())/(y.max()-y.min())
#considering 90% oftraining data (i.e,first 3 values)
length= (90*len(x))//100
print(length)
train_x = x[:length]
train_y = y[:length]
#considering 10% of testing data (i.e,last value)

test_x = x[length:]
test_y = y[length:]
print(test_y)
m=1;
c=1;
eta=0.01;
epochs=10000;
for i in range(epochs):
  tm = 0;
  tc=0;
  for j in range(len(train_x)):
    p =m*train_x[j]+c
    temp= train_y[j]-p
    tm += temp*train_x[j]
    tc += temp
  gdm = -(tm)/len(train_x)
  gdc = -(tc)/len(train_x)
  deltam = -eta*gdm
  deltac = -eta*gdc
  m = m + deltam
  c = c + deltac

print(m,c)

q = 0
q1=0
for i in range(len(train_x)):
  v=m*train_x[i]+c
  z = int(train_y[i])-int(v)
  q += z**2
  q1+=abs(z)

mse_train = q/len(train_x)
mae_train = q1/len(train_x)
print("mean square error for tarining data:",mse_train)
print("Root mean square error for training data:",mse_train**0.5)
print("mean absolute error for trained dataset:",mae_train)



# calculating Mean square error for training data set
w =0
w1=0
for i in range(len(test_x)):
  v=m*train_x[i]+c
  z = int(train_y[i])-int(v)
  w += z**2
  w1 +=abs(z)

mse_test = w/len(test_x)
mae_test = w1/len(test_x)
print("mean square error for test data:",mse_test)
print("Root mean square error for testing data:",mse_test**0.5)
print("mean absolute error for testing data:",mae_test)
#denormalising dataset
for i in range(len(x)):
  x[i] = x[i]*(x_max - x_min)+x_min
  y[i] = y[i]*(y_max - y_min)+y_min

print(y)

# Deploying model  
x_value =float(input("Enter value of x:"))
x_value = (x_value-x_min)/(x_max-x_min)
y_value = m*x_value+c
y_value = y_value*(y_max-y_min)+y_min

print("y value when x value is",x_value,": ",y_value)
```
