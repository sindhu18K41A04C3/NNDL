```
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
d = pd.read_csv('/content/Active power load dataset.csv')
print(d)
x_min = d['X'].min()
x_max = d['X'].max()
y_min = d['Y'].min()
y_max = d['Y'].max()

#normalisation
d=(d-d.min())/(d.max()-d.min())
print(d)

x=np.array(d.iloc[:,0:1])
y=np.array(d.iloc[:,1:2])



#spliting training and testing data
train_x,test_x = train_test_split(x,test_size =0.1)
train_y,test_y= train_test_split(y,test_size =0.1)

m=1
c=1
eta=0.01
epochs=10000
gama = 0.9
vm,vc =0,0

for i in range(epochs):
  for j in range(len(train_x)):
    temp= train_y[j]-(m*train_x[j])-c
    gdm = -train_x[j]*temp
    gdc = -temp
    deltam = -eta*gdm
    deltac = -eta*gdc
    vm = gama*vm +deltam
    vc = gama*vc +deltac
    m = m + vm
    c = c + vc
#printing final model parametres(m and c) values  
print("Final updated model parametres m = ",m,"and c =",c)

# calculating Mean square error,Root mean square error and mean absolute error for training data set
q = 0
q1=0
for i in range(len(train_x)):
  v=m*train_x[i]+c
  z = int(train_y[i])-int(v)
  q += z**2
  q1+=abs(z)

mse_train = q/len(train_x)
mae_train = q1/len(train_x)
print("mean square error for tarining data:",mse_train)
print("Root mean square error for training data:",mse_train**0.5)
print("mean absolute error for trained dataset:",mae_train)


#calculating Mean square error,Root mean square error and mean absolute error for testing data set
w =0
w1=0
for i in range(len(test_x)):
  v=m*train_x[i]+c
  z = int(train_y[i])-int(v)
  w += z**2
  w1 +=abs(z)

mse_test = w/len(test_x)
mae_test = w1/len(test_x)
print("mean square error for test data:",mse_test)
print("Root mean square error for testing data:",mse_test**0.5)
print("mean absolute error for testing data:",mae_test)
for i in range(len(x)):
  x[i] = x[i]*(x_max - x_min)+x_min
  y[i] = y[i]*(y_max - y_min)+y_min

print(y)
# Deploying model  
x_value =float(input("Enter value of x:"))
x_value = (x_value-x_min)/(x_max-x_min)
y_value = m*x_value+c
y_value = y_value*(y_max-y_min)+y_min

print("Estimated load based on its previous value is: ",y_value)
```
