```
import numpy as np
import pandas as pd
d = pd.read_csv('/content/Active power load dataset.csv')
print(d)
x=np.array(d.iloc[:,1:2])
y=np.array(d.iloc[:,2:3])
x_min = x.min()
x_max = x.max()
y_min = y.min()
y_max = y.max()
#Normalising the dataset values
for i in range(len(x)):
  x[i] = (x[i]-x.min())/(x.max()-x.min())
  y[i] = (y[i]-y.min())/(y.max()-y.min())

print(y)
#considering 90% oftraining data (i.e,first 81 values)
length= (90*len(x))//100
print(length)
train_x = x[:length]
train_y = y[:length]
#considering 10% of testing data (i.e, next 9 value)

test_x = x[length:]
test_y = y[length:]
print(len(train_x))
print(len(test_y))
#Considering values for model parametres(m,c),eta and no of iterations (epochs),gama,vm and vc
m=1
c=1
eta=0.001
epochs=100000
gama = 0.9
vm,vc =0,0

#calculating gradient desent for model parametres(m and c) 
#step size(deltam and deltac) 
#updating model parametres
for i in range(epochs):
  for j in range(len(train_x)):
    temp= train_y[j]-((m+(gama*vm))*train_x[j])-(c+(gama*vc))
    gdm = -train_x[j]*temp
    gdc = -temp
    deltam = -eta*gdm
    deltac = -eta*gdc
    vm = gama*vm +deltam
    vc = gama*vc +deltac
    m = m + vm
    c = c + vc
#printing final m and c values  
print("Final m value is: ",m,"Final c value is: ",c)

# calculating Mean square error, Root mean square error, mean absolute error for training dataset
q = 0
q1=0
for i in range(len(train_x)):
  v=m*train_x[i]+c
  z = int(train_y[i])-int(v)
  q += z**2
  q1 +=abs(z)

mse_train = q/len(train_x)
mae_train = q1/len(train_x)
print("mean square error for tarined data:",mse_train)
print("Root mean square error for tarined data:",mse_train**0.5)
print("mean absolute error for trained dataset:",mae_train)


# calculating Mean square error, Root mean square error, mean absolute error for testing dataset
w =0
w1=0
for i in range(len(test_x)):
  v=m*train_x[i]+c
  z = int(train_y[i])-int(v)
  w += z**2
  w1 += abs(z)


mse_test = w/len(test_x)
mae_test = w1/len(test_x)
print("mean square error for testing data:",mse_test)
print("Root mean square error for testing data:",mse_test**0.5)
print("mean absolute error for testing data:",mae_test)

# Denormalising the data
for i in range(len(x)):
  x[i] = ((x[i]*(x_max-x_min))+x_min)
  y[i] = ((y[i]*(y_max-y_min))+y_min)
print(y)

# Deploying model to predict data
x_value = float(input("Enter loat at previous hour: "))
y_value = (m*x_value)+c
print("Load at particular hour based on load at previous hour is : ",y_value)
```
